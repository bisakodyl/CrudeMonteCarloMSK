```python
import numpy as np
import pandas as pd
import math
import random
from matplotlib import pyplot as plt, figure
plt.style.use('dark_background')
from IPython.display import clear_output
pi = 3.1415926
```

```python
def get_rand_number(min,max):
    range = max - min
    choice = random.uniform(0,1)
    return min + range*choice
```

```python
def f_of_x(x):
    return (1/(pi*math.sqrt(x*(1-x))))
```

```python
def crude_monte_carlo(num_of_samples,min,max):
    sum_of_samples = 0
    data = []
    for i in range(num_of_samples):
        x = get_rand_number(min,max)
        y = f_of_x(x)
        sum_of_samples += y
        data.append([x,y])
    return ((max - min) * float(sum_of_samples/num_of_samples)), data
```

```python
approx,data = crude_monte_carlo(10,0,1)
print("Monte Carlo Approximation of f(x) for 10 samples: ", approx)
df = pd.DataFrame(data, columns=['x','y'])
df.sort_values(by='x', ascending=True,inplace=True)

#ten drop tutaj po to żeby wykres był ładny, bo przyblizenie cos nie dziala
df.drop(df.loc[df['y']>3].index, inplace=True)
plt.figure(figsize=(20, 10))
plt.hist(df['y'], density=True,bins = 10)
plt.xlabel('y values')
plt.ylabel('frequency')


pd.DataFrame({'x': df['x'],
              'y': df['y']}).plot(x='x',y='y',label='Monte Carlo Approximation of f(x) for 10 samples',
       figsize=(5,10));
```
