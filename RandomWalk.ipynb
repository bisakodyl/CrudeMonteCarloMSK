```python
def get_rand_number(min,max):
    range = max - min
    choice = random.uniform(0,1)
    return min + range*choice

def zerolistmaker(n):
    listofzeros = [0,0] * n
    return listofzeros
```

```python
step_n = 10
steps = []

begin = array([[int(get_rand_number(min_start,max_start)),int(get_rand_number(min_start,max_start))]])
begin = begin/1

for i in range(step_n):
    step = np.random.choice([-1,1])
    direction = np.random.choice([-1,1])
    if (direction == 1):
        steps.append([0,step])

    if (direction == -1):
        steps.append([step,0])
        
path = np.concatenate([begin, steps]).cumsum(0)



#--------------------------------------------------------
start = path[:1]
stop = path[-1:]

fig = plt.figure(figsize=(8,8),dpi=200)
ax = fig.add_subplot(111)

#zakrety
ax.scatter(path[:,0], path[:,1],c='yellow',alpha=0.25,s=1);
#droga
ax.plot(path[:,0], path[:,1],c='blue',alpha=0.5,lw=1,ls='-');
#start
ax.plot(start[:,0], start[:,1],c='green', marker='o')
#stop
ax.plot(stop[:,0], stop[:,1],c='red', marker='o')

plt.title('Random Walk Mareczka: ' + str(step_n) + ' krokow')
plt.tight_layout(pad=0)
```
